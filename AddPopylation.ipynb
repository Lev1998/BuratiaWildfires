```python
def add_nasel(df):
  for ind,row in df.iterrows():
    pop=int(city[city['OBJECTID *']==row['id_city']]['population'])
    df.loc[ind,'POPYLATION']=pop
```

```python
def replaceSign(df):
  for i,row in df.iterrows():
    df.loc[i,'latitude']=str(row['latitude']).replace(',','.')
    df.loc[i,'longitude']=str(row['longitude']).replace(',','.')
    df.loc[i,'Road_DIST']=str(row['Road_DIST']).replace(',','.')
    df.loc[i,'Railway_DIST']=str(row['Railway_DIST']).replace(',','.')
    df.loc[i,'City_DIST']=float(str(row['City_DIST']).replace(',','.'))
    df.loc[i,'NDVI']=str(row['NDVI']).replace(',','.')
```

```python
dfm=[]
for i in range(0,4):
  data=pd.read_csv('/content/drive/My Drive/ForestFire/MODIS_Point/Final/таблицы без населения(NDVI)/MN'+str(i+2017)+'NDVI.csv',sep=';')
  data['POPYLATION']=0
  add_nasel(data)
  replaceSign(data)
  data.to_csv('/content/drive/My Drive/ForestFire/MODIS_Point/Final/ForTraining(NDVI)/ModisComplite'+str(i+2017)+'.csv',index=False)
  dfm.append(data)
```

```python
dataset=pd.concat([dfm[0],dfm[1],dfm[2],dfm[3]],ignore_index=True)
dataset=dataset[dataset['LandClass']!='<Null>']
dataset=dataset[dataset['NDVI']!='<Null>']
```
