```python
fpm2017=pd.read_csv('/content/drive/My Drive/ForestFire/FirePointsGroup/MODIS/FireBurModis2017.csv')
fpm2018=pd.read_csv('/content/drive/My Drive/ForestFire/FirePointsGroup/MODIS/FireBurModis2018.csv')
fpm2019=pd.read_csv('/content/drive/My Drive/ForestFire/FirePointsGroup/MODIS/FireBurModis2019.csv')
fpm2020=pd.read_csv('/content/drive/My Drive/ForestFire/FirePointsGroup/MODIS/FireBurModis2020.csv')
fpm2021=pd.read_csv('/content/drive/My Drive/ForestFire/FirePointsGroup/MODIS/FireBurModis2021.csv')
dataList=[fpm2017,fpm2018,fpm2019,fpm2020,fpm2021]

nameNew=['nFireBurModis2017','nFireBurModis2018','nFireBurModis2019','nFireBurModis2020','nFireBurModis2021']
```

```python
#Добавление атрибута дата и время
def chislo(chis):
  if chis<10:
    itog='0'+str(chis)
  else:
    itog=str(chis)
  return itog
def addDatatime(df):
  for index,row in df.iterrows():
    time=chislo(row['acq_time']//100)+':'+chislo(row['acq_time']%100)
    date=str(row['acq_date']).replace('.','/')
    df.loc[index,'time']=date + ' ' + time
```

```python
ind=0
for i in dataList:
  addDatatime(i)
  i.to_csv('/content/drive/My Drive/ForestFire/FirePointsGroup/MODIS/'+nameNew[ind]+'.csv',index=True)
  ind+=1
```
