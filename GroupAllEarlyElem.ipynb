```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
def firstElem(df):
  df=df[['latitude','longitude','acq_date','GROUP_ID','acq_time','daynight']]
  df = df.sort_values(by='GROUP_ID')
  df_S=pd.DataFrame()
  for i in df['GROUP_ID'].unique():
    df_p=df[df['GROUP_ID']==i]
    df_p.sort_values(by='acq_date')
    r=df_p.iloc[0]
    df_p=df_p[df_p['acq_date']==r['acq_date']]
    df_p=df_p.sort_values(by='acq_time')
    #print(df_p)  
    if df_p.shape[0]>1:
      for i in range(0,df_p.shape[0]-1):
        df_S=df_S.append(df_p.iloc[i], ignore_index=True)
        if(not((df_p.iloc[i]['acq_date']==df_p.iloc[i+1]['acq_date'])and(df_p.iloc[i]['acq_time']==df_p.iloc[i+1]['acq_time']))):
          break
    else:
      df_S=df_S.append(df_p.iloc[0], ignore_index=True)

  return df_S
```

```python
for i in range(0,5):
  fir_dfO=pd.read_csv('/content/drive/My Drive/ForestFire/FirePointsGroup/'+str(i+2017)+'/modGroup'+str(i+2017)+'.csv',sep=';')
  rez=firstElem(fir_dfO)
  rez.to_csv('/content/drive/My Drive/ForestFire/FirePointsGroup/'+str(i+2017)+'/MAFP'+str(i+2017)+'.csv',index=True)
```
