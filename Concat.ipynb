```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
def replaceSign(df):
  for i,row in df.iterrows():
    df.loc[i,'latitude']=str(row['latitude']).replace(',','.')
    df.loc[i,'longitude']=str(row['longitude']).replace(',','.')
```

```python
for i in range(0,5):
  fir_dfA=pd.read_csv('/content/drive/My Drive/ForestFire/FirePointsGroup/'+str(i+2017)+'/MAFP'+str(i+2017)+'.csv')
  fir_dfA=fir_dfA[['longitude','latitude']]
  fir_dfA['fire']=1
  rand_dfA=pd.read_csv('/content/drive/My Drive/ForestFire/MODIS_Point/ModisRandomPoint/RandomAllPoint/MRP'+str(i+2017)+'All.csv',sep=';')
  rand_dfA['fire']=0
  rezA=pd.concat([fir_dfA,rand_dfA],ignore_index=True)
  replaceSign(rezA)
  rezA.to_csv('/content/drive/My Drive/ForestFire/MODIS_Point/Final/'+str(i+2017)+'AllforData.csv',index=False)
```
